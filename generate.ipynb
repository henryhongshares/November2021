```python
from PIL import Image 
from IPython.display import display 
import random
import json
import time
```

```python
# Each image is made up a series of traits
# The weightings for each trait drive the rarity and add up to 100%


background_textures = ["Green","Grey", "Light Blue", "Green Red", "Pink Blue", "Pink", "Purple", "Yellow"] 
background_textures_weights = [8 ,8 ,8 ,8 ,8 ,8 , 8, 8]


meta_cat = ["Blue", "Dark Pink", "Green", "Grey", "Lavender","Light Blue", "Orange", "Pink", "Purple Pink", "White Blue","White Red", "Yellow"] 
meta_cat_weights = [6 ,6 ,6 ,6 ,6 ,6 ,6 ,6 ,6 ,6 ,6 ,6 ]

hand_and_tail = ["Black Heart", "Black Star"] 
hand_and_tail_weights = [50, 50]

spacesuit = ["Green Red", "Pink Blue", "White Blue", "White Green", "White Grey", "White Pink", "White Purple", "White Yellow"] 
spacesuit_weights = [12.5, 12.5, 12.5, 12.5, 12.5, 12.5, 12.5, 12.5,]

# Dictionary variable for each trait. 
# Eech trait corresponds to its file name

background_textures_files = {
    "Green": "green",
    "Grey": "grey",
    "Light Blue": "light_blue",
    "Green Red": "mix_ green_red",
    "Pink Blue": "mix_pink_blue",
    "Pink": "pink",
    "Purple": "purple",
    "Yellow": "yellow",

}


meta_cat_files = {
    "Blue": "blue",
    "Dark Pink": "dark_pink",
    "Green": "green",
    "Grey": "grey",
    "Lavender": "lavender",
    "Light Blue": "light_blue",
    "Orange": "orange",
    "Pink": "pink",
    "Purple Pink": "purple_pink",
    "White Blue": "white_blue",
    "White Red": "white_red",
    "Yellow": "yellow",
}

hand_and_tail_files = {
    "Black Heart": "black_heart",
    "Black Star": "black_star",
    
}

spacesuit_files = {
    "Green Red": "mix_green_red",
    "Pink Blue": "mix_pink_blue",
    "White Blue": "white_blue",
    "White Green": "white_green",
    "White Grey": "white_grey",
    "White Pink": "white_pink",
    "White Purple": "white_purple",
    "White Yellow": "white_yellow",

}

```

```python
## Generate Traits

TOTAL_IMAGES = 50 # Number of random unique images we want to generate

all_images = [] 

# A recursive function to generate unique image combinations
def create_new_image():
    
    new_image = {} #

    # For each trait category, select a random trait based on the weightings 
    new_image ["Background_Textures"] = random.choices(background_textures, background_textures_weights)[0]
    new_image ["Meta_Cat"] = random.choices(meta_cat, meta_cat_weights)[0]
    new_image ["Hand_and_Tail"] = random.choices(hand_and_tail, hand_and_tail_weights)[0]
    new_image ["Spacesuit"] = random.choices(spacesuit, spacesuit_weights)[0]
    
    if new_image in all_images:
        return create_new_image()
    else:
        return new_image
    
    
# Generate the unique combinations based on trait weightings
for i in range(TOTAL_IMAGES): 
    
    new_trait_image = create_new_image()
    
    all_images.append(new_trait_image)
    
```

```python
# Returns true if all images are unique
def all_images_unique(all_images):
    seen = list()
    return not any(i in seen or seen.append(i) for i in all_images)

print("Are all images unique?", all_images_unique(all_images))
```

```python
# Add token Id to each image
i = 0
for item in all_images:
    item["tokenId"] = i
    i = i + 1
```

```python
# Get Trait Counts

background_textures_count = {}
for item in background_textures:
    background_textures_count[item] = 0

meta_cat_count = {}
for item in meta_cat:
    meta_cat_count[item] = 0

hand_and_tail_count = {}
for item in hand_and_tail:
    hand_and_tail_count[item] = 0

spacesuit_count = {}
for item in spacesuit:
    spacesuit_count[item] = 0    
    
    
for image in all_images:
    meta_cat_count[image["Meta_Cat"]] += 1
    background_textures_count[image["Background_Textures"]] += 1
    hand_and_tail_count[image["Hand_and_Tail"]] += 1
    spacesuit_count[image["Spacesuit"]] += 1
    
print(meta_cat_count)
print(background_textures_count)
print(hand_and_tail_count)
print(spacesuit_count)
```

```python
#### Generate Metadata for all Traits 
METADATA_FILE_NAME = './metadata/all-traits.json'; 
with open(METADATA_FILE_NAME, 'w') as outfile:
    json.dump(all_images, outfile, indent=4)
```

```python
start_time = time.time()
spacesuit

#### Generate Images in the following order   
for item in all_images:
    
    im1 = Image.open(f'./trait-layers/2_background_textures/{background_textures_files[item["Background_Textures"]]}.png').convert('RGBA')
    im2 = Image.open(f'./trait-layers/3_meta_cats_folder/{meta_cat_files[item["Meta_Cat"]]}.png').convert('RGBA')
    im3 = Image.open(f'./trait-layers/4_hand_and_tail/{hand_and_tail_files[item["Hand_and_Tail"]]}.png').convert('RGBA')
    im4 = Image.open(f'./trait-layers/5_spacesuit/{spacesuit_files[item["Spacesuit"]]}.png').convert('RGBA')

    #Create each composite
    com1 = Image.alpha_composite(im1, im2)
    com2 = Image.alpha_composite(com1, im3)
    com3 = Image.alpha_composite(com2, im4)
    
    
    #Convert to RGB (make sure to specify which com e.g com1 or com2)
    rgb_im = com3.convert('RGB')
    file_name = str(item["tokenId"]) + ".png"
    rgb_im.save("./images/" + file_name)
    
##############################################################################################################################    
runtime = time.time() - start_time
runtime_whole_minutes = round(runtime/60)
runtime_leftover_minutes= runtime % 60

print (f"My program took, {runtime} seconds, to run")
print (f"which is, {runtime_whole_minutes}:{runtime_leftover_minutes} minutes, to run for {TOTAL_IMAGES} images")
```

```python
#### Generate Metadata for each Image    

f = open('./metadata/all-traits.json',) 
data = json.load(f)


IMAGES_BASE_URI = "ADD_IMAGES_BASE_URI_HERE"
PROJECT_NAME = "ADD_PROJECT_NAME_HERE"

def getAttribute(key, value):
    return {
        "trait_type": key,
        "value": value
    }
for i in data:
    token_id = i['tokenId']
    token = {
        "image": IMAGES_BASE_URI + str(token_id) + '.png',
        "tokenId": token_id,
        "name": PROJECT_NAME + ' ' + str(token_id),
        "attributes": []
    }
    token["attributes"].append(getAttribute("Meta_Cat", i["Meta_Cat"]))
    token["attributes"].append(getAttribute("Background_Textures", i["Background_Textures"]))
    token["attributes"].append(getAttribute("Hand_and_Tail", i["Hand_and_Tail"]))
    token["attributes"].append(getAttribute("Spacesuit", i["Spacesuit"]))
    
    with open('./metadata/' + str(token_id), 'w') as outfile:
        json.dump(token, outfile, indent=4)
f.close()
```
